```python
# Calculating ODE for 1 dimensional system
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt

#Define growth parameters
a= 0.66 #growth rate of prey
b= 1.33 #rate at which predetor destroy prey
c= 1.0#death rate of predetor
d= 1.0 #the rate at which predetors increases by consuming prey
t0=0

#setting intial condition
p0=1.0 #intial condition for prey
q0=1.0 #intial condition for predator
y0= [p0,q0] #initial condition for y 

#Define model(function) for the population growth for 2 dimensions
def growth(y,t):
    p,q=y
    dpdt= a*p-b*p*q
    dqdt= -c*q+d*p*q
    dydt= [dpdt,dqdt]
    return dydt

#define time scale 
t = np.linspace(t0,40, 200)

#apply odeint
sol=odeint(growth,y0,t)

#Plot the graph 
plt . figure(figsize=(15,7))
plt.plot(t,sol[:,0],'-b',label='Prey')
plt.plot(t,sol[:,1],'-r',label='Predator')
plt.plot(sol[:,0],sol[:,1])
plt . legend ( loc='best', fontsize =18)
plt.xlabel("Time",fontsize=18)
plt.ylabel("No. of species",fontsize=18)
plt.grid(True)
plt.show()

```

```python
#Creating twin plots
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(15,10))

#First grah -  A plot of the solutions p(t) and q(t) for t ∈ [0, 50]
ax1.plot(t,sol[:,0],'-b',label='Prey')
ax1.plot(t,sol[:,1],'-r',label='Predator')
ax1.legend ( loc='best', fontsize =18)
ax1.set_xlabel("Time",fontsize=18)
ax1.set_ylabel("No. of species",fontsize=18)
ax1.grid(True)

#Phase plot
ax2.plot(sol[:,0],sol[:,1])
ax2.set_xlabel("Prey(x)",fontsize=18)
ax2.set_ylabel("Predator(y)",fontsize=18)
```

```python
#Plotting one curve for each initial condition
fig = plt.figure()
fig.set_size_inches(6,6) # Square plot, 1:1 aspect ratio
ics = np.arange(1.0, 3.0, 0.1) # initial conditions
for r in ics:
    y0 = [r, 1.0]
    sol = odeint(growth, y0, t)
    plt.plot(sol[:,0], sol[:,1], "-")
plt.xlabel("Rabbits", fontsize=14)
plt.ylabel("Foxes", fontsize=14)
```
